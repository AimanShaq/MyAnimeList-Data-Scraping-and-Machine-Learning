```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dfanime = pd.read_csv('anime.csv')
dfrating = pd.read_csv('rating.csv')

dfanime

dfanime.isnull().sum()
```

```python
dfanime = dfanime.dropna()
dfanime.isnull().sum()
```

```python
i = dfanime[dfanime.episodes=='Unknown'].index
dfanime = dfanime.drop(i)

dfanime
```

```python
dfmerge = pd.merge(dfanime, dfrating, on=['anime_id','anime_id'])
dfmerge = dfmerge.rename(columns={'rating_x':'rating_anime','rating_y':'rating_user'})
dfmerge
```

```python
allgenres = ''
for genre in dfanime['genre'].to_list():
    allgenres += str(genre) + ', '
listgenres = allgenres.split(', ')
allgenres = set(listgenres)
allgenres.remove('')

allgenres
```

```python
rating_count = dfanime.groupby(dfanime['type'])['rating']
rating_count.agg([('Minimum','min'),('Maximum','max'),('Average','mean')])
```

```python
member_count = dfmerge.groupby(dfmerge['type'])['members']
member_count.agg([('Minimum','min'),('Maximum','max'),('Average','mean')])
```

```python
genres_count = []
for genre in allgenres:
        gen = [genre,listgenres.count(genre)]
        genres_count.append(gen)

genredf = pd.DataFrame(genres_count,columns=['Genre','Count'])
genredf.set_index('Genre',inplace=True)
genredf
```

```python


fig = plt.figure(figsize=(16,8))
topgenredf = genredf.sort_values(by=['Count'],ascending=False)[:10]
topgenredf.plot.bar()
```
